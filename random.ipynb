```python
# Online Python compiler (interpreter) to run Python online.
# Write Python 3 code in this online editor and run it.
d = {}
with open("coding_qual_input.txt") as f:
    for line in f:
        (key, val) = line.split()
        d[int(key)] = val
    print(f.read())
increasing = 1
addUp = 1
buildResult=""
while addUp < len(d):
    buildResult = buildResult+" "+d[addUp]
    increasing += 1
    addUp += increasing
if addUp is not len(d):
    buildResult = buildResult+" "+d[len(d)]

print(buildResult)
```
